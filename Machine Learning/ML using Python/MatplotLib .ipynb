### Scatter plot

```python
plt.scatter(x,y,c='g')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Graph in 2D')
plt.savefig("fig.png")
```

```python
y=x*x
```

```python
plt.plot(x,y,'r*-',linestyle='dashed',linewidth=2,markersize=12)
```

```python
#creating subplot

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'r')
```

```python
x=np.arange(1,11)
y=3*x+5
plt.plot(x,y)
```

```python
np.pi
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave form")
plt.plot(x,y)
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,2,1)
plt.plot(x,y_sin,'r--')
plt.title('sine wave')

plt.subplot(2,2,2)
plt.plot(x,y_cos,'g--')
plt.title('cos wave')
```

### Bar Plot

```python
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y)
plt.bar(x2,y2,color='g')
```

### Histogram

```python
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a,bins=20)
plt.show()
```

### Box Plot

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)
plt.show()
```

### Pie Chart

```python
labels='Python','C++','Ruby','Java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,
       autopct="%1.1f%%",shadow=True)
plt.axis('equal')
plt.show()
```
